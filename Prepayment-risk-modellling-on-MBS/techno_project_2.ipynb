```python
#find number of different element in each column

def count_unique_elements(df):
  """
  This function returns a dictionary containing the number of unique elements in each column of a dataframe.

  Args:
      df: The pandas dataframe.

  Returns:
      A dictionary with the column names as keys and the number of unique elements as values.
  """

  unique_counts = {}
  for column in df.columns:
    unique_counts[column] = df[column].nunique()
  return unique_counts

unique_counts = count_unique_elements(df)

for column, count in unique_counts.items():
  print(f"Column: {column}, Unique elements: {count}")
```
